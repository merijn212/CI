```python
import pandas as pd
import json

def load_jsons(data_path, file):
    """ helper function to load '.json' files (they're not proper jsons) """
    file_path = data_path + file
    with open(file_path) as jsons:
        lines = [json.loads(json_line) for json_line in jsons]
    return pd.DataFrame(lines)
```

```python
# Load data for town called Ambridge
data_path = './data/ambridge/'
review_file = 'review.json'
business_file = 'business.json'
user_file = 'user.json'
tip_file = 'tip.json'
checkin_file = 'checkin.json'

reviews = load_jsons(data_path, review_file)
businesses = load_jsons(data_path, business_file)
users = load_jsons(data_path, user_file)
tips = load_jsons(data_path, tip_file)
checkins = load_jsons(data_path, checkin_file)
```

```python
def pivot_ratings(ratings):
    return ratings.pivot_table(index = "user_id", columns = "business_id", values="stars")
utility_matrix = pivot_ratings(reviews)


utility_matrix = utility_matrix.replace(0, np.nan)
display(utility_matrix)
```

```python
import pandas as pd
import json
import sklearn.metrics.pairwise as pw
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
def mean_center_rows(matrix):
    return matrix.sub(matrix.mean(axis=1), axis=0)
mean_matrix = mean_center_rows(utility_matrix)
def create_similarity_matrix_cosine(matrix):
    """Creates a adjusted(/soft) cosine similarity matrix.

    Arguments:
    matrix -- a utility matrix

    Notes:
    Missing values are set to 0. This is technically not a 100% correct, but is more convenient 
    for computation and does not have a big effect on the outcome.
    """
    mc_matrix = matrix - matrix.mean(axis = 0)
    return pd.DataFrame(pw.cosine_similarity(mc_matrix.fillna(0)), index = matrix.index, columns = matrix.index)
similarity = create_similarity_matrix_cosine(mean_matrix)
similarity = similarity.replace(0, np.nan)
display(similarity.head())
```

```python
def select_neighborhood(similarities, ratings, k):

    list_ = []
    ratings = ratings.reindex(ratings.dropna().index)
    
    for rating in ratings.index:
        list_.append(rating)
    
    similarities = similarities.loc[list_]

    similarities = similarities[similarities > 0]
    similarities = similarities.sort_values(ascending=False)
    similarities = similarities[:k]
    return similarities

neighborhood1 = select_neighborhood(similarity["01vXmAJKQ41PVi2wgYDiPA"], utility_matrix["muFJIZKZwbAfy_pEFKF_pw"], 50)

display(neighborhood1)
```

```python
def weighted_mean(neighborhood, ratings): 

    s = neighborhood.mul(ratings, fill_value=np.nan)
    s = s.reindex(s.dropna().index)
    noemer = sum(s)
    teller = sum(neighborhood)

    if noemer == 0 or teller == 0:
        return np.nan

    return noemer/teller

prediction1 = weighted_mean(neighborhood1, utility_matrix["muFJIZKZwbAfy_pEFKF_pw"])
print(prediction1)
```

```python
def mse(predicted_ratings):
    """Computes the mean square error between actual ratings and predicted ratings
    
    Arguments:
    predicted_ratings -- a dataFrame containing the columns rating and predicted rating
    """
    diff = predicted_ratings['rating'] - predicted_ratings['predicted rating']
    return (diff**2).mean()
```

```python
list_predict = []
for i, row in reviews.iterrows():
        neighbors = select_neighborhood(similarity[row["user_id"]],utility_matrix[row["business_id"]], 50)
        list_predict.append(weighted_mean(neighbors, utility_matrix[row["business_id"]] ))
reviews["predicted rating"] = list_predict
reviews = reviews.drop(["useful", "funny", "cool", "text", "date"], axis=1)
display(reviews)
```
